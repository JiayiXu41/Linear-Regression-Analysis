### Key Attributes

**ample_id**: A unique identifier assigned to each banana sample in the dataset.

**variety**: The cultivar or breed of banana, such as Cavendish, Red Dacca, or Lady Finger. 

**region**: The geographic origin of the banana, such as Ecuador, Philippines, or Costa Rica. 

**quality_score**: A numerical score, likely on a scale of 1-4 that rates the overall quality of the banana sample. 

**quality_category**: A text label that categorizes the quality score into broader groupings like "Excellent" etc.

**ripeness_index**: A numerical index representing the ripeness level of the banana, potentially ranging from 1 (green/unripe) to 10 (overripe). This quantifies the maturity of the fruit.

**ripeness_category**: A text label like "Green", "Yellow", "Ripe", or "Overripe" that corresponds to the ripeness index. This gives a clear, qualitative ripeness classification.

**sugar_content_brix**: The sugar content of the banana measured in degrees Brix. 

**firmness_kgf**: The firmness of the banana measured in kilograms-force. 

**length_cm**: The physical length of the banana in centimeters.

### Question

- **Calculate the correlation coefficient of quality scores and sugar content.**  I plan to use df.corr() to build a correlation matrix. From the section 3, I got the scatter plot showing the relationship between quality scores and sugar content. So I want to calculate the accurate correlation coefficient to verify my conclusion. If it is closer to -1 or 1, it means that sugar content predicts quality scores well. In this case, I will build a model in the next question.

- **Use the machine learning model to predict the quality_score using sugar_content_brix.** If the correlation coefficient is as expected, I will build a model to predict quality score. The question will be **"what is the quality prediction for a 19 brix sugar content banana?"** This question applies machine learning techniques. I plan to fit the linear regression model with the data at first, than I plan to use the model to predict quality score.

### Solution to Q1

```python
df.corr(numeric_only=True)
```

- ***From the matrix, I found that the correlation coefficient of quality scores and sugar content is 0.602842. It shows a moderate to strong positive relationship between quality scores and sugar content. So I can build a model in the next part.***

### Solution to Q2

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
ind = ['sugar_content_brix']
dep = 'quality_score'
model.fit(df[ind], df[dep])
model.predict([[19]])
```

- ***So a banana with 19 brix sugar content is predicted to have 2.54258885 quality score.***
